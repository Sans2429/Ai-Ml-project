```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
rd=pd.read_csv('/content/50_Startups.csv')
rd.head()
```

```python
rd.isnull().sum()
```

```python
plt.plot(rd['Marketing Spend'],rd['Profit'],color='b',label='mp')
plt.scatter(rd['Marketing Spend'],rd['Profit'],color='r',marker='*')
plt.plot(rd['R&D Spend'],rd['Profit'],color='yellow',label='rp')
plt.scatter(rd['R&D Spend'],rd['Profit'],color='g',marker='o')
plt.legend()
plt.show()
```

```python
l=rd['State'].unique()
st=rd.value_counts('State')
plt.bar(l,st)
```

```python
c=[75,15,10]
e=[0.3,0,0]
plt.pie(c,labels=rd['State'].unique(),explode=e)
plt.show()
```

```python
sns.pairplot(data=rd,hue='State')
```

```python
x1=rd['Marketing Spend']
y1=rd['Profit']
sns.histplot(data=rd,x=x1,y=y1,hue='State')
```

```python
y2=rd['State']
sns.boxplot(data=rd,x=x1,y=y2)
```
